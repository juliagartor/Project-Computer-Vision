```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('n.bmp')
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

template = img[500:555,970:1090]
#template = cv.cvtColor(template, cv.COLOR_BGR2GRAY)

img_rgb = cv.imread('n.bmp')
img_gray = cv.cvtColor(img_rgb, cv.COLOR_BGR2GRAY)

#img_rgb = cv.imread('test.bmp')
#img_gray = cv.cvtColor(img_rgb, cv.COLOR_BGR2GRAY)

w, h = template.shape[::-1]

res = cv.matchTemplate(img_gray,template,cv.TM_CCOEFF_NORMED)
threshold = 0.8

loc = np.where( res >= threshold)

for pt in zip(*loc[::-1]):
    cv.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (0,0,255), 2)

plt.imshow(img_rgb)
```

```python
r = img_rgb[pt[1]-8:pt[1]+h, pt[0]+w+5:pt[0]+w+270]
cv.imwrite("number.png",r)
```

```python
canny = cv.Canny(r, 0,250, 4)
plt.imshow(canny, cmap='gray')
```

```python
(cnt, hierarchy) = cv.findContours(
    canny.copy(), cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)
```
